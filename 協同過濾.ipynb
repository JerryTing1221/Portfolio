```python
import pandas as pd
import numpy as np
import glob,os
```

### 先將所有 excel表 匯成 DataFrame

```python
#存放檔案路徑
path = '/Users/Jerry/Desktop/Time series/google map'

file = glob.glob(os.path.join(path, "*.xlsx"))

dl= []

for value in file:
    dl.append(pd.read_excel(value,usecols = ['使用者','地點','評價(1~5)'],index_col=None))

Total_df =pd.concat(dl,axis=0)

Total_df.dropna()

Total_df
```

### 把評價轉換成數字型態

```python
rating = []

for i in Total_df['評價(1~5)']:
    i = i.replace(' 顆星','')
    i = i.replace(' ','')
    if i == '':
        i = np.nan
    i = float(i)
    rating.append(i)

Total_df['評價(1~5)'] = rating
Total_df['評價(1~5)']
```

### 去除空值

```python
Total_df.dropna()
```

### 把表的型態變成pivot table

```python
pivot_table = Total_df.pivot_table(index=['使用者'],columns=['地點'],values="評價(1~5)")

pivot_table
```

### 使用函數corrwith做協同過濾

```python
point = pivot_table["赤崁樓"]
similarity_with_other = pivot_table.corrwith(point)  
similarity_with_other = similarity_with_other.sort_values(ascending=False)
similarity_with_other.head()
```
