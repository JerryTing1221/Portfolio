## googlemap 協同過濾

```python
import pandas as pd
import numpy as np
import glob,os
```

```python
#存放檔案路徑
path = './googlemap'

file = glob.glob(os.path.join(path, "*.xlsx"))

dp= []

for value in file:
    dp.append(pd.read_excel(value,usecols = ['使用者','地點','評價(1~5)'],index_col=None,engine='openpyxl'))

Total_df =pd.concat(dp,axis=0)

Total_df.dropna()

Total_df
```

```python
#評價轉數值型態

rating = []

for i in Total_df['評價(1~5)']:
    i = i.replace(' 顆星','')
    i = i.replace(' ','')
    if i == '':
        i = np.nan
    i = float(i)
    rating.append(i)

Total_df['評價(1~5)'] = rating
Total_df['評價(1~5)']

```

```python
Total_df.dropna()

pivot_table = Total_df.pivot_table(index=['使用者'],columns=['地點'],values="評價(1~5)")
pivot_table = pivot_table.rename(columns={'B.B.ART':'B_B_ART'})  #mongodb的key不可含'.'or'$'
pivot_table
# pivot_table.to_excel('./協同過濾.xlsx',encoding="utf-8")
```

## word2vec model 存取

```python
import pickle   #將python 序列化和反序列化的模塊  pickling=>dump儲存
from pymongo import MongoClient
from gensim.models import word2vec
model = word2vec.Word2Vec.load("./word2vec (1).model")
# print(model)
# type(model)
pickled_wvmodel = pickle.dumps(model)
print(pickled_wvmodel)
model_name = 'wvmodel'
client = MongoClient("mongodb://" + "ec2-3-16-42-183.us-east-2.compute.amazonaws.com"+":27017") # No private key  
auth = client.admin.authenticate("chi","101") # Authenticate to admin
db = client['travel_project']
collection = db['models']
collection.insert_one({model_name:pickled_wvmodel,'name':model_name})
```

```python
json_data = {}
data = collection.find({'name': model_name}) 
for i in data: 
    json_data = i #fetching model from db 
pickled_model = json_data[model_name] 
pickle.loads(pickled_model)

```
